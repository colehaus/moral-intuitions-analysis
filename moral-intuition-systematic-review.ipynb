```python
import numpy as np
from scipy import stats

# https://www.frontiersin.org/articles/10.3389/fpsyg.2013.00863/full

# d

# Between-subjects

# Hedges' correction multiplied or divided?
def hedges_g(m1, m2, n1, n2, s1, s2):
    return (m1 - m2) / (pooled_std_dev(n1, n2, s1, s2) * hedges_correction(n1, n2))

def hedges_correction(n1, n2):
    return 1 - 3/(4 * (n1 + n2) - 9)

def pooled_std_dev(n1, n2, s1, s2):
    return np.sqrt(((n1 - 1) * s1 ** 2 + (n2 - 1) * s2 ** 2) / (n1 + n2 - 2))

def d_s_from_t(t, n1, n2):
    return t * np.sqrt(1/n1 + 1/n2)

def r_from_d_s(d, n1, n2):
    N = n1 + n2
    return d / np.sqrt(d ** 2 + (N ** 2 - 2 * N) / (n1 * n2))

# Within-subjects

def S_diff(SD1, SD2, r):
    return np.sqrt(SD1 ** 2 + SD2 ** 2 - 2 * r * SD1 * SD2)

# Rarely used in meta-analysis
def cohens_d_z(M_diff, S_diff):
    return M_diff / S_diff

# Rarely used in meta-analysis
def cohens_d_z_from_t(t, n):
    return t / np.sqrt(n)

# Not generally recommended
def cohens_d_rm(M_diff, S_diff, r):
    return M_diff / S_diff * np.sqrt(2 * (1 - r))

# Generally recommended
def cohens_d_av(M_diff, SD1, SD2):
    return M_diff / ((SD1 + SD2) / 2)

# r

# between and within

def eta_p_squared_from_f(F, df_effect, df_error):
    return (F * df_effect) / (F * df_effect + df_error)

# within-subjects designs that control for intra-subjects variability (η2p and ω2p), 
# or that take the correlation between measurements into account (Cohen's dz) are 
# an accurate description of the size of the effect, or 
# whether effect sizes that do not control for intra-subjects variability (η2G and ω2G), or 
# that control for correlation between measurements (e.g., Cohen's drm or Cohen's dav) are preferred

# https://sci-hub.tw/10.1146/annurev.psych.52.1.59

def eta_from_chi_squared(chi_squared, df, n):
    if type(n) == list:
        n_tot = sum(n)
    else:
        n_tot = n
    assert(df == 1)
    return np.sqrt(chi_squared / n_tot)

def d_from_r(r):
    return 2 * r / np.sqrt(1 - r ** 2)
    
def eta_from_d(d):
    return np.sqrt(d ** 2 / (d ** 2 + 4))

# assert np.abs(r_from_d(d_from_r(.8)) - 0.8) < 0.01
# assert np.abs(d_from_r(r_from_d(.8)) - 0.8) < 0.01

def eta_from_z(z, n):
    return z / np.sqrt(n)

def eta_from_t(t, df):
    return np.sqrt(t ** 2 / (t ** 2 + df))

def z_from_p(p):
    return np.abs(stats.norm.ppf(p))

# Me

def eta_squared_from_r(r):
    return r ** 2
```

```python
import inspect

def filter_for_required_args(fn, d):
    required_args = set(inspect.signature(fn).parameters)
    return {k: v for k, v in d.items() if k in required_args}

def call_with_only_required_args(fn, d):
    return fn(**filter_for_required_args(fn, d))

def standardize(experiment):
    try: 
        experiment["partial_eta_squared"]
    except KeyError:
        try:
            experiment["partial_eta_squared"] = eta_squared_from_r(experiment["r"])
        except:
            try:
                experiment["partial_eta_squared"] = eta_p_squared_from_f(F = experiment["f"][1], df_effect = experiment["f"][0][0], df_error = experiment["f"][0][1])
            except:
                try:
                    experiment["partial_eta_squared"] = call_with_only_required_args(eta_from_chi_squared, experiment) ** 2
                except:
                    try:
                        experiment["partial_eta_squared"] = call_with_only_required_args(eta_from_d, experiment) ** 2
                    except:
                        try:
                            experiment["partial_eta_squared"] = eta_from_t(t = experiment["t"][1], df = experiment["t"][0]) ** 2
                        except TypeError as t:
                            print(experiment)
                            raise t
            
    return experiment

experiments_standardized = list(map(standardize, experiments))
experiments_standardized
```

```python
import seaborn as sns
from pandas import DataFrame
from matplotlib import pyplot as plt

sns.set(style="darkgrid")

def forest_plot(name, data, figsize):
    fig = plt.figure(tight_layout=True, figsize=figsize)
    sns.pointplot(x="partial_eta_squared", y="study", data=data, join=False)
    fig.patch.set_alpha(0)
    plt.savefig(name)

def funnel_plot(name, data, figsize):
    fig = plt.figure(tight_layout=True, figsize=figsize)
    sns.scatterplot(x="partial_eta_squared", y="n", data=data, legend=False)
    fig.patch.set_alpha(0)
    plt.savefig(name)

clean_disgust = DataFrame([{
    "partial_eta_squared": e["partial_eta_squared"],
    "study": e["ref"] + " " + e["experiment"],
    "n": sum(e["n"])
} for e in experiments_standardized if e["predictor"][0] == "clean" or e["predictor"][0] == "disgust"])
    
    
forest_plot("clean-forest.png", clean_disgust, (9, 4))
funnel_plot("clean-funnel.png", clean_disgust, (9, 4))

order = DataFrame([{
    "partial_eta_squared": e["partial_eta_squared"],
    "study": e["ref"] + " " + e["experiment"],
    "n": sum(e["n"])
} for e in experiments_standardized if e["predictor"][0] == "order"])

forest_plot("order-forest.png", order, (9, 4))
funnel_plot("order-funnel.png", order, (9, 4))
```
